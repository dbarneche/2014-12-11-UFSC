```python
from ipythonblocks import ImageGrid
import random
SIZE = 19

# Make a size X size grid.
grid = []
for i in range(SIZE):
    grid.append([0] * SIZE)
grid[(SIZE + 1) / 2][0] = 255

# Inject particles until the aggregate reaches the edge.
while True:
    
    # Start the particle at the top.
    x = random.randint(0, SIZE-1)
    y = SIZE-1
    
    # Move the particle until it hits something.
    while True:
        # Check for neighbors.
        free = True
        if (x > 0) and grid[x-1][y]:
            free = False
        if (x < SIZE-1) and grid[x+1][y]:
            free = False
        if (y > 0) and grid[x][y-1]:
            free = False
        # Adjacent to something, so fill in and quit loop.
        if not free:
            grid[x][y] = 255
            break

        # If it didn't stick, move it.
        y -= 1
        direction = random.randint(0, 2)
        if  direction == 0:
            x = max(0, x-1)
        elif direction == 2:
            x = min(SIZE-1, x+1)
        
        # Fallen off the bottom?
        if y < 0:
            break
    
    # Check if the last particle was at the top.
    if y == SIZE-1:
        break

# Draw the final grid.
display = ImageGrid(SIZE, SIZE, fill=(200, 200, 200))
for x in range(SIZE):
    for y in range(SIZE):
        if grid[x][y]:
            display[x, y] = (grid[x][y], 0, 0)
display.show()
```
